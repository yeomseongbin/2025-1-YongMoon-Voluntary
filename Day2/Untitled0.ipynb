```python
import torch
import torch.nn.functional as F

# 입력 데이터와 정답
x = torch.tensor([[0.5, 0.8, 0.2]], dtype=torch.float32)
y = torch.tensor([[1.0]], dtype=torch.float32)

# 가중치 및 편향 초기화
w1 = torch.tensor([[1.0, -1.0],
                   [0.0, 2.0],
                   [1.0, 0.5]], requires_grad=True)
b1 = torch.tensor([[0.1, 0.2]], dtype=torch.float32, requires_grad=True)

w2 = torch.tensor([[2.0],
                   [-1.0]], dtype=torch.float32, requires_grad=True)
b2 = torch.tensor([[0.5]], dtype=torch.float32, requires_grad=True)

# 학습 설정
lr = 0.1
epochs = 10

print("--- 학습 시작 ---")
for epoch in range(1, epochs + 1):
    # 순전파
    z1 = x @ w1 + b1
    a1 = F.relu(z1)
    z2 = a1 @ w2 + b2
    y_pred = F.relu(z2)
    loss = 0.5 * (y_pred - y) ** 2

    # 역전파
    loss.backward()

    # 파라미터 업데이트
    with torch.no_grad():
        w1 -= lr * w1.grad
        b1 -= lr * b1.grad
        w2 -= lr * w2.grad
        b2 -= lr * b2.grad

        # 기울기 초기화
        w1.grad.zero_()
        b1.grad.zero_()
        w2.grad.zero_()
        b2.grad.zero_()

    print(f"Epoch {epoch} | y_pred: {y_pred.item():.4f} | loss: {loss.item():.4f}")
```
